```python
import requests
from bs4 import BeautifulSoup

rating_pages = []
pages = range(1, 100)
for page in pages:
    url = "https://movie.naver.com/movie/point/af/list.naver?&page={}".format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'html.parser')
    for i in soup.find_all("td", {"class":"title"}):
      rating_pages.append(i.get_text())

movies = []
for item in rating_pages:
  item_mod = item.replace("\t","")
  movies.append(item_mod)


semi_final = []
for item in movies:
  item_mod = item.replace("\n"," ")
  semi_final.append(item_mod)


final = []
for item in semi_final:
  item_mod = item.replace("신고 ","")
  final.append(item_mod)
```

```python
final
```

```python
## 영화 이름 추출
titles = []
for item in final:
  item_title = item.split("  별점")
  title = item_title[0]
  titles.append(title[1:])
titles
```

```python
##영화 별점 추출
numbers = []
for item in final:
  results = item.split(" ")
  num_str = results[results.index("별점")+4]
  num = num_str[1:]
  numbers.append(num)
numbers
```

```python
#설명추출
reviews = []
for item in final:
  results = item.split("별점")
  review_str = results[1].lstrip('10점')
  reviews.append(review_str[14:])
reviews
```

```python
## titles, numbers, reviews
import pandas as pd
df = pd.DataFrame()

df['titles'] = titles
df['rates'] = numbers
df['reviews'] =reviews

df
```
