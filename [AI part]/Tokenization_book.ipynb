```python
#모델 생성
from konlpy.tag import Komoran
komoran = Komoran()
```

```python
import pandas as pd
import numpy as np

#영화 리뷰데이터 가져오기
reviews = pd.read_csv('/content/drive/MyDrive/book_reviews(descriptions,img_url).csv')
reviews
```

```python
print(reviews.iloc[0][4]) #review
print(reviews.iloc[0][5]) #description
```

```python
import re
no_eng = []
for i in range(0, reviews.shape[0]):
  text = reviews.iloc[i][4]
  text = re.sub('[a-zA-Z]' , '', text)
  emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
  text = emoji_pattern.sub(r'', text)
  no_eng.append(text)
no_eng[3495]
```

```python
lists = []

#토큰화 진행
for i in range(0, reviews.shape[0]):
  if i == 3495:
    mini_1 = komoran.pos('일상에 늘 있지만 어렵게 생각되는 철학을쉽고 짧게 풀어 놓아서 철학 입문서로 적당한 듯 하다.그러나 철학서는 쉬우나 어려우나 딱딱한 빵이니 오래 씹어야 한다.')
    mini_2 = []
    for k in mini_1:
      if k[1] in ('NNG', 'NNP', 'MM', 'MAG', 'IC', 'NF', 'NV'):
        mini_2.append(k[0])
      elif k[1] in ('VV', 'VA'):
        mini_2.append(k[0]+"다")
  else :
    mini_1 = komoran.pos(no_eng[i])
    mini_2 = []
    for k in mini_1:
      if k[1] in ('NNG', 'NNP', 'MM', 'MAG', 'IC', 'NF', 'NV'):
        mini_2.append(k[0])
      elif k[1] in ('VV', 'VA'):
        mini_2.append(k[0]+"다")
  print(i)
  print(mini_2)
  lists.append(mini_2)

#리스트에 저장 후 영화리뷰데이터에 별도 열로 추가
reviews['komoran'] = lists
```
